```python
# урок взят с https://smart-lab.ru/blog/592975.php
# Сначала испытаем yfianance  пакет. Его можно установить при помощи команды pip install yfinance. 
# Import the yfinance. If you get module not found error the run !pip install yfinance from your Jupyter notebook
import yfinance as yf
import pandas as pd
tickers_list = tickers_list = ['ALRS.ME','VTBR.ME','GAZP.ME','GAZA.ME','LNTA.ME',
                               'NVTK.ME','SBER.ME','SNGS.ME','TATN.ME','YNDX.ME']

# Import pandas
data = pd.DataFrame(columns=tickers_list)

# Fetch the data

for ticker in tickers_list:
    data[ticker] = yf.download(ticker,'2016-01-01')['Adj Close']

# Print first 5 rows of the data
data.head()
```

```python
# Import the plotting library
import matplotlib.pyplot as plt
%matplotlib inline
#  При помощи пакета  matplotlib и полученных данных можно построить график дневной доходности.
# Plot all the close prices
# Функция pct_change () удобна для использования, но в свою очередь скрывает то, как получается значение. Схожее вычисление, которое поможет понять механизм, можно выполнить при помощи shift() из пакета из пакета Pandas. Дневная цена закрытия делится на прошлую (сдвинутую на один) цену и из полученного значения вычитается единица. Но есть один незначительный минус – первое значение в результате получается NA.
# cumprod() вычисляет кумулятивную доходность прибавляет предыдущий результат
((data.pct_change()+1).cumprod()).plot(figsize=(10, 7))

# Show the legend
plt.legend()

# Define the label for the title of the figure
plt.title("Скорректированая цена закрытия", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Year', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
```
